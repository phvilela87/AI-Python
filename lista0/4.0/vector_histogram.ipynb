```python
"""
Histogram plot
"""

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

M = 10000

# Load variables X and Y, both are uniform distributed and transpose to column vectors
x = np.random.uniform(size=(1, M)).T
y = np.random.uniform(size=(1, M)).T

z = x + y

plt.title('Histogram')
plt.hist(z, bins=1000, density=True, color='g')

plt.savefig('vector_histogram.png')
```
