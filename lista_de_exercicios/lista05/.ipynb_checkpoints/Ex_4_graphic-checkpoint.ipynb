```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(start=0, stop=5, num=100)
y = 5 - x
```

```python
fig = plt.figure()
plt.plot(x, y, linewidth=2, label='Decision Boundary')
plt.text(1, 2, 'y = 0', fontsize=16)
plt.text(3, 3, 'y = 1', fontsize=16)
plt.legend()
plt.grid()
plt.savefig("graphic.png", dpi=600)
plt.show()
```
