```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=5, cluster_std=1.0, shuffle=True, random_state=42)
```

```python
idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)
idx2 = np.argwhere(y == 2)
idx3 = np.argwhere(y == 3)
idx4 = np.argwhere(y == 4)

fig = plt.figure(figsize=(5,5))
plt.plot(X[idx0,0], X[idx0,1], '.', label='Class 0')
plt.plot(X[idx1,0], X[idx1,1], '>', label='Class 1')
plt.plot(X[idx2,0], X[idx2,1], 'o', label='Class 2')
plt.plot(X[idx3,0], X[idx3,1], 'v', label='Class 3')
plt.plot(X[idx4,0], X[idx4,1], '^', label='Class 4')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('a)')
plt.legend()
plt.show()
```

```python
km = KMeans(n_clusters=4)
km.fit(X)
```

```python
centroid1 = []
centroid2 = []

print('c) valores ótimos para centroids')

for i in range(0,4):
    centroid1.append(km.cluster_centers_[i][0])
    centroid2.append(km.cluster_centers_[i][1])
    print('Centroid %d: (%1.2f, %1.2f)' % (i,km.cluster_centers_[i][0],km.cluster_centers_[i][1]))
```

```python
print('Foram necessárias',km.n_iter_, 'Iterações para treinar o modelo')
```

```python
y = km.labels_
```

```python
idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)
idx2 = np.argwhere(y == 2)
idx3 = np.argwhere(y == 3)
idx4 = np.argwhere(y == 4)

fig = plt.figure(figsize=(5,5))
plt.plot(X[idx0,0], X[idx0,1], '.', label='Class 0')
plt.plot(X[idx1,0], X[idx1,1], '>', label='Class 1')
plt.plot(X[idx2,0], X[idx2,1], 'o', label='Class 2')
plt.plot(X[idx3,0], X[idx3,1], 'v', label='Class 3')
plt.plot(X[idx4,0], X[idx4,1], '^', label='Class 4')
plt.plot(centroid1,centroid2, 'P', label='Centroids')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('e) clusters a que cada um pertence')
plt.legend()
plt.show()
```
