```python
"""
2D graphic plot
"""

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Number of elements
M = 1000

# Load variables X and W, where X is uniform and W normal distribution and transpose to column vectors
x = np.random.uniform(size=M).T
w = np.random.randn(M).T

y = (1.2 + (2.3 * x) + (10 * w)).T

# plotting points as a scatter plot
plt.scatter(x, y, color="gray")
plt.title('2D Graphic')

plt.savefig('graphic_2D.png')
```
