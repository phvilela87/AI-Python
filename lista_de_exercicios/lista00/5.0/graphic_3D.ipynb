```python
"""
3D graphic
"""

import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

x1 = np.arange(start=-10, stop=11, step=0.25)
x2 = np.arange(start=-10, stop=11, step=0.25)

y = np.power(x1, 2) + np.power(x2, 2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x1, x2, y, 'blue')

ax.view_init(34, -11)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')

plt.savefig('graphic_3D.png')
```
