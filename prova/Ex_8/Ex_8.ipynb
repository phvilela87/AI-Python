```python
import numpy as np
import tensorflow.compat.v1 as tf
```

```python
tf.disable_v2_behavior()
```

```python
# to make this notebook's output stable across runs
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

## Gradient Descent Optimizer

```python
reset_graph()
```

```python
print (tf.__version__)
 
# input X vector
X = [[0, 0], [0, 1], [1, 0], [1, 1]]
# output Y vector
Y = [[0], [1], [1], [0]]
 
# Placeholders for input and output
x = tf.placeholder("float", shape=[4,2])
y = tf.placeholder("float", shape=[4,1])
 
# W matrix
W1 = tf.Variable([[1.0, 0.0], [1.0, 0.0]], shape=[2,2])
W2 = tf.Variable([[0.0], [1.0]], shape=[2,1])
 
# Biases
B1 = tf.Variable([0.0, 0.0], shape=[2])
B2 = tf.Variable([0.0], shape=1)
 
# Hidden layer and outout layer
output =tf.sigmoid(tf.matmul(tf.sigmoid(tf.matmul(x, W1) + B1), W2) + B2)
 
# error estimation
e = tf.reduce_mean(tf.squared_difference(y, output))
train = tf.train.GradientDescentOptimizer(0.1).minimize(e)
 
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
 
for i in range (40001):
    error = sess.run(train, feed_dict={x: X, y: Y})
    if i % 100 == 0:
        print('\nEpoch: ' + str(i))
        print('\nError: ' + str(sess.run(e, feed_dict={x: X, y: Y})))
        for el in sess.run(output, feed_dict={x: X, y: Y}):
            print('    ',el)
sess.close()
 
print ("Complete")
```

```python
# x and y are placeholders
x = tf.placeholder("float")
y = tf.placeholder("float")
 
# Equation y = b*x*x
b = tf.Variable([0.0], name="b")
ys = tf.multiply(tf.multiply(x, x),b)
 
# Error 
e = tf.square(y - ys)
 
# Gradient Descent Optimizer
# 1.0 is learning_rate. The learning rate to use by Tensorflow.
train = tf.train.GradientDescentOptimizer(1.0).minimize(e)
 
# Create a session and run the model
model = tf.global_variables_initializer()
 
with tf.Session() as session:
    session.run(model)
    for i in range(50):
        x_value = np.random.rand()
        # Assume b=4, then let's find b_value = 4 and find it 
  # by gradient descent
        # y = 4*x*x
        y_value = 4.0*x_value*x_value
        session.run(train, feed_dict={x: x_value, y: y_value})
 
    b_value = session.run(b)
    print("Predicted model: b = ", b_value)
```

## Adam Optimizer

```python
reset_graph()
```

```python
print (tf.__version__)
 
# input X vector
X = [[0, 0], [0, 1], [1, 0], [1, 1]]
# output Y vector
Y = [[0], [1], [1], [0]]
 
# Placeholders for input and output
x = tf.placeholder("float", shape=[4,2])
y = tf.placeholder("float", shape=[4,1])
 
# W matrix
W1 = tf.Variable([[1.0, 0.0], [1.0, 0.0]], shape=[2,2])
W2 = tf.Variable([[0.0], [1.0]], shape=[2,1])
 
# Biases
B1 = tf.Variable([0.0, 0.0], shape=[2])
B2 = tf.Variable([0.0], shape=1)
 
# Hidden layer and outout layer
output =tf.sigmoid(tf.matmul(tf.sigmoid(tf.matmul(x, W1) + B1), W2) + B2)
 
# error estimation
e = tf.reduce_mean(tf.squared_difference(y, output))
train = tf.train.AdamOptimizer(0.1).minimize(e)
 
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
 
for i in range (40001):
    error = sess.run(train, feed_dict={x: X, y: Y})
    if i % 100 == 0:
        print('\nEpoch: ' + str(i))
        print('\nError: ' + str(sess.run(e, feed_dict={x: X, y: Y})))
        for el in sess.run(output, feed_dict={x: X, y: Y}):
            print('    ',el)
sess.close()
 
print ("Complete")
```

```python
# x and y are placeholders
x = tf.placeholder("float")
y = tf.placeholder("float")
 
# Equation y = b*x*x
b = tf.Variable([0.0], name="b")
ys = tf.multiply(tf.multiply(x, x),b)
 
# Error 
e = tf.square(y - ys)
 
# Adam Optimizer
# 0.1 is learning_rate. The learning rate to use by Tensorflow.
train = tf.train.AdamOptimizer(0.1).minimize(e)
 
# Create a session and run the model
model = tf.global_variables_initializer()
 
with tf.Session() as session:
    session.run(model)
    for i in range(50):
        x_value = np.random.rand()
        # Assume b=4, then let's find b_value = 4 and find it 
  # by gradient descent
        # y = 4*x*x
        y_value = 4.0*x_value*x_value
        session.run(train, feed_dict={x: x_value, y: y_value})
 
    b_value = session.run(b)
    print("Predicted model: b = ", b_value)
```
